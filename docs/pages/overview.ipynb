**snippet-cast** turns an annotated Python snippet into a narrated screencast
video. Narration is written as trailing `#:` comments on the snippet's own
lines, so the input stays valid, runnable Python. The tool renders
syntax-highlighted code frames with a progressive reveal, a live variable
state panel, optional burned-in captions and a typing-in animation,
synthesises speech per line, and stitches everything into an MP4 with ffmpeg.


## Installation

::: {.panel-tabset}

```bash
pixi workspace channel add munch-group
pixi add snippet-cast
```

## conda

```bash
conda install -c munch-group snippet-cast
```

## pip

```bash
pip install snippet-cast
```

:::


## Basic use

```bash
snippet-cast snippet.py -o out.mp4 --tts silent --subtitles
```

## Demo

`--every` animates a loop iteration by iteration instead of revealing the
snippet line by line, rendered here with `--tts silent --subtitles`:

```bash
snippet-cast loop.py -o loop_every.mp4 --every --subtitles --tts silent
```
