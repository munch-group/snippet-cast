```python
%%snippet-cast --tts manual --record -n forl
#%%snippet-cast --tts say -n forl
#: 1) Lets compute the Fibonacci numbers / Let's trace an iterative Fibonacci function as it runs.
def fib(n):             #: 3) Now define the function. / The first line names the fib-function and defines its parameters.
    a, b = 0, 1         #: 4) "a" and "b" are our first two Fibonacci numbers / Start from the first two Fibonacci numbers, zero and one.
    for _ in range(n):  #: 5) / Loop n times;`` the counter itself is unused.
        a, b = b, a + b #: 6) / Advance the pair — b becomes the running sum.
    return a            #: 7) / After the loop, a holds the nth Fibonacci number.

result = fib(7)         #: 2) We call our function "fib" and make it take a single argument / Call fib with seven; result becomes thirteen.
```
